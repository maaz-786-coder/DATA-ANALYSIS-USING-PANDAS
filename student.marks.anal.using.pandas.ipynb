```python

import pandas as pd
import numpy as np

# Create sample dataset
data = {
    "Name": ["Ali", "Sara", "John", "Aisha", "Rahul"],
    "Math": [85, 90, 78, 92, 88],
    "Science": [80, 85, 75, 95, 82],
    "English": [78, 88, 70, 90,  85]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Convert to numpy array (only marks)
marks = df[["Math", "Science", "English"]].to_numpy()

# Calculate average using numpy
avg_marks = np.mean(marks, axis=1)

df["Average"] = avg_marks
print(df)
```

```python
top_student = df.loc[df["Average"].idxmax()]
print(top_student)
subject_avg = df[["Math", "Science", "English"]].mean()
print(subject_avg)
```

```python
# Students with average > 85
high_performers = df[df["Average"] > 85]
print(high_performers)
def grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    else:
        return "C"

df["Grade"] = df["Average"].apply(grade)
print(df)
```
